```python
import layoutparser as lp
import cv2
```

```python

image = cv2.imread("image_1.png")
image = image[..., ::-1] 
```

```python
model = lp.Detectron2LayoutModel('lp://PubLayNet/mask_rcnn_X_101_32x8d_FPN_3x/config',extra_config=["MODEL.ROI_HEADS.SCORE_THRESH_TEST", 0.75],
                                 label_map={0: "Text", 1: "Title", 2: "List", 3:"Table", 4:"Figure"})
```

```python

layout = model.detect(image)
```

```python
lp.draw_box(image, layout, box_width=3,show_element_id=True)
```
